```python
#Read csv file provided
output_data = "output_data/cities.csv"

# Define latitude and Longitute, I took the values from the world map lat and long
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
#Define empty lists for cities and long and lat

lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
```

```python
#retrieve news cities, create empty lists

new_cities = []
cloudiness = []
country = []
date = []
humidity = []
temperature = []
lat = []
lng = []
wind_speedy = []
```

```python
cloudiness
```

```python
# Build partial query URL
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="
```

```python
#Testing url
url
```

```python
#testing weather_api_key

weather_api_key
```

```python
#looping into the cities
#using enumerate to get the index of the list

for i, city in enumerate(cities):
               
    try:
        print(i,city)
        response = requests.get(query_url + city).json()
        lat.append(response['coord']['lat'])
        temperature.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speedy.append(response['wind']['speed'])
        lng.append(response['coord']['lon'])
        new_cities.append(city)
            
    except Exception as e: #e however problem has in the list and ir tells you what the error is
        print(e)    

new_cities        
```

```python
requests.get(query_url + city).json()
```

```python
#Dataframe

weather_df = pd.DataFrame({"cities": new_cities,
              "cloudiness": cloudiness,
              "humidity":humidity,
              "Temp": temperature,
              "lat": lat,
              "long":lng,
              "windy speed":wind_speedy 
})
weather_df.head()
weather_df.to_csv("weather.csv")
```

```python
# Create Northern and Southern Hemisphere DataFrames
mask = weather_df['lat'] > 0
northern_hemisphere = weather_df[mask]
southern_hemisphere = weather_df[mask]
```

```python
x_values = northern_hemisphere["lat"]
y_values = northern_hemisphere["Temp"]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,0),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temp')
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
print(f"The r-squared is: {rvalue**2}")
plt.savefig('output_data/nh_lat_temp_regression.png')
```
